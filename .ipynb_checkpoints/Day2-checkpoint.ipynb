```python
with open("orig_en_train.json") as f:
    data = json.load(f)
    # print("number of tweets", len(data))
```

```python
tweet_counts = defaultdict(int)
tweet_likes = defaultdict(int)
for d in data:
    tweet_counts[d['tweet_time'][:10]] += 1
    if int(d['like_count']) > 0:
        tweet_likes[d['tweet_time'][:10]] += 1
sorted_X = sorted(tweet_counts.items(), key=operator.itemgetter(0))
sorted_Y = sorted(tweet_likes.items(), key=operator.itemgetter(0))
```

```python
n_tweets = defaultdict(int)
followers = defaultdict(int)
hs_users = defaultdict(set)
all_hs = defaultdict(int)
hs_likes = defaultdict(int)
for d in data:
    curr_hashtags = d['hashtags']
    curr_hashtags = curr_hashtags[1:-1].split(', ')
    for hs in curr_hashtags:
        if hs!="":
            all_hs[hs] += 1
    if int(d['like_count']) > 0:
        for hs in curr_hashtags:
            if hs!="":
                hs_likes[hs] += 1
    hs_users[d['userid']].update(curr_hashtags)
    n_tweets[d['userid']] += 1
    followers[d['userid']] = int(d['follower_count'])
```
